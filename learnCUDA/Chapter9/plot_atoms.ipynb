```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

## Plot points from xy.txt

```python
xy = np.loadtxt('xy.txt')
x = xy[:, 0]
y = xy[:, 1]
plt.figure(figsize=(10, 10))
plt.scatter(x, y, s=0.5, c='r')
plt.axis('equal')
plt.axis('off')
plt.title('Atoms without bonds')
plt.style.use('default')
plt.show()
```
